# Linear Regression Project Part 3 : Converting Categorical data into Numerical Data 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Python\Linear_Reg_Project_data\Ames_NO_Missing_Data.csv", keep_default_na = False)
```

```python
pd.set_option('display.max_rows', None)
df.isnull().sum()
```

## I have observed that the MS SubClass column is not a ordinal numercial type, and hence I will convert it to string

```python
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

```python
# Listing down columns that are of object data type
my_object_df = df.select_dtypes(include = 'object')
my_numeric_df = df.select_dtypes(exclude = 'object')
```

```python
# Converting them to numeric using get dummies - dtype = int will show 0 and 1 instead of True and False
df_object_dummies = pd.get_dummies(my_object_df, drop_first = True, dtype = int)
```

```python
df_object_dummies.head()
```

```python
final_df = pd.concat([my_numeric_df, df_object_dummies], axis = 1)
```

```python
final_df.head()
```

```python
final_df.corr()['SalePrice'].sort_values(ascending = False)
```

```python
final_df.to_csv(r"D:\Python\Linear_Reg_Project_data\AMES_Final_DF.csv")
```
